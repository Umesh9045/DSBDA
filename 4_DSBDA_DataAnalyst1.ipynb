```python
#Split dependent variable and independent variables
x = df.drop('medv', axis=1)
y = df.medv
```

```python
#spliting the data to traning and testing dataset
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3)
```

```python
# Use linear regression( Train the Machine ) to Create Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
# Predict the y_predict for all values of train_x and test_x
y_predict = lr.predict(xtrain)
y_test_pred = lr.predict(xtest)
```

```python
# Calculate mean square eror for train_y and test_y
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(ytest, y_test_pred)
```

```python
#Plotting linear regression model
plt.scatter(ytrain ,y_predict,c='blue',marker='o',label='Training data')
plt.scatter(ytest,y_test_pred ,c='lightgreen',marker='s',label='Test data')
plt.xlabel('True values')
plt.ylabel('Predicted')
plt.title("True value vs Predicted value")
plt.legend(loc= 'upper left')
plt.plot()
plt.show()
```

**Linear Regression** is statistical method that can be used if relationship between dependent variable and independent variable can be approximated within a straight line.
It is used to predict continuous dependent variable.

**Mean Square Error (MSE)** is a widely used metric to measure the average squared difference between the predicted values and the actual values in a regression problem. It is a common evaluation metric for assessing the quality or accuracy of a predictive model.

**MSE = (1/n) * Σ(yi - ŷi)^2**

where:

* MSE is the Mean Square Error
* n is the total number of data points
* yi is the actual value of the target variable for the ith data point
* ŷi is the predicted value of the target variable for the ith data point
* Σ denotes the summation over all data points

**lr.fit(xtrain, ytrain)**

The **fit() method** adjusts the parameters of the model based on the input data to find the best fit or the optimal coefficients.
